1 Computing with Language: Texts and Words

```python
# Searching Text
print("\n",text1)
text1.concordance("monstrous")

print("\n",text2)
text2.concordance("monstrous")
```

```python
# A concordance permits us to see words in context
print("\n",text1)
text1.similar("monstrous")
print("\n",text2)
text2.similar("monstrous")
```

```python
# The term common_contexts allows us to examine just the contexts that are shared by two or more words,
print("\n",text1)
text1.common_contexts(["monstrous", "very"])
print("\n",text2)
text2.common_contexts(["monstrous", "very"])
```

```python
# determine the location of a word in the text
text4.dispersion_plot(["citizens", "democracy", "freedom", "duties", "America"])
text2.dispersion_plot(["father",'love','sister','mother','hate'])
```

```python
book_temp = text2
print(book_temp)
print('Words:','\t'*2,len(book_temp))
print('Unique Words:','\t',len(set(book_temp)))
print('Share','\t'*2,np.round(len(set(book_temp))/len(book_temp),2))
```

```python
def lexical_diversity(text:str):
    return len(set(text)) / len(text)
def percentage(count, total):
    return 100 * count / total
```

```python
# tells us the frequency of each vocabulary item in the text
fdist1 = FreqDist(text1)
print(fdist1.most_common(50))
fdist1['whale']
```

```python
# Cumulative Frequency Plot for N Most Frequently
N = 10
fdist1.plot(N, cumulative=True)
```

```python
# look at the long words of a text
V = set(text2)
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```

```python
#  Long and frequentally often words
fdist1 = FreqDist(text1)
sorted(w for w in set(text1) if len(w) > 10 and fdist1[w] > 50)
```
